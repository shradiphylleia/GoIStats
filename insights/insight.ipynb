```python
# rural mpce,urban mpce state disparity
x = np.arange(len(df['name'])) * 1.3
width=0.5

fig,ax=plt.subplots(layout='constrained',figsize=(15,8))
bars_rural=ax.bar(x-width/2,df['mpce_rural'],width,label='MPCE Rural')
bars_urban=ax.bar(x+width/2,df['mpce_urban'],width,label='MPCE Urban')

for bars in [bars_rural,bars_urban]:
    for bar in bars:
        height=bar.get_height()

        ax.set_xlabel('State/UT/India')
        ax.set_ylabel('MPCE')
        ax.set_title('MPCE in states by rural and urban')
        ax.set_xticks(x)
        ax.set_xticklabels(name,rotation=90)
        ax.legend()

plt.savefig('rural_urban_mpce.png')
```

```python
from mpce_product_category import data_rural
mpce_products_rural=pd.DataFrame(data_rural)
```

```python
mpce_products_rural.head(20)
```

```python
mpce_products_rural['Category'].value_counts()
```

```python
def product_category_viz():
    for i in range(0, 32):
        category_name=mpce_products_rural.iloc[i]["Category"]
        food_product_state=mpce_products_rural.iloc[i].to_frame(name='spend')

        food_product_state = food_product_state.drop(index="Category", errors='ignore')

        mean=food_product_state['spend'].mean()
        median=food_product_state['spend'].median()

        name=food_product_state.index  

        fig,ax=plt.subplots(figsize=(14, 6))
        ax.bar(name,food_product_state['spend'], label='State Spend')

        ax.axhline(y=mean, xmin=0.02, xmax=0.99, linewidth=1.5,color='g', marker="*", label='Mean', linestyle=':')
        ax.axhline(y=median, linewidth=2, marker='*', linestyle=':',color='m',label='Median')

        ax.set_ylabel('Product MPCE')
        ax.set_xlabel('State/UT/India')
        ax.set_title(f'MPCE Distribution for Product: {category_name}')

        plt.xticks(rotation=90)

        ax.legend(loc='upper right', fontsize=12)
        plt.show()

product_category_viz()
```
