```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
```

```python
root_dir = os.path.abspath("..")
sys.path.append(root_dir)

from mpce_social_grp import data_8R
df_sg=pd.DataFrame(data_8R)
df_sg.head()
```

```python
df_sg=df_sg.fillna(0)
```

```python
df_sg.isna().sum()
```

```python
import seaborn as sns
```

```python
df_sg_melt=df_sg.melt(id_vars=['name'],var_name='Category', value_name='Value')
```

```python
plt.figure(figsize=(22,8))

sns.barplot(x="name", y="Value", hue="Category", data=df_sg_melt)

plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("MPCE Value(Rs.)")
plt.title("Grouped Bar Plot of Different Categories by State")
plt.legend(title="Category")

plt.savefig('social_grp.png')
```
