```python
def product_category_viz():
    for i in range(0, 32):
        category_name=mpce_products_rural.iloc[i]["Category"]
        food_product_state=mpce_products_rural.iloc[i].to_frame(name='spend')

        food_product_state = food_product_state.drop(index="Category", errors='ignore')

        mean=food_product_state['spend'].mean()
        median=food_product_state['spend'].median()

        name=food_product_state.index  

        fig,ax=plt.subplots(figsize=(14, 6))
        ax.bar(name,food_product_state['spend'], label='State Spend')

        ax.axhline(y=mean, xmin=0.02, xmax=0.99, linewidth=1.5,color='g', marker="*", label='Mean', linestyle=':')
        ax.axhline(y=median, linewidth=2, marker='*', linestyle=':',color='m',label='Median')

        ax.set_ylabel('Product MPCE')
        ax.set_xlabel('State/UT/India')
        ax.set_title(f'MPCE Distribution for Product: {category_name}')

        plt.xticks(rotation=90)

        ax.legend(loc='upper right', fontsize=12)
        plt.show()

        fig.savefig(f'MPCE_{category_name}.png', bbox_inches='tight')

product_category_viz()
```
